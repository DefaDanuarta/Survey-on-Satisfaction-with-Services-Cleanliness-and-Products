# Data Preprocessing

```python
likert_map = {
        'sangat tidak setuju':1,
        'tidak setuju':2,
        'tidak merasakan':3,
        'setuju':4,
        'sangat setuju':5
}
freq_map = {
        '0 kali': 0,
        '1-2 kali': 1,
        '3-4 kali': 2,
        '>4 kali': 3
}
bool_map = {
        'iya': 1,
        'tidak': 0
}
```

```python
df_clean = df.copy()
exclude_cols = ['Nama Lengkap Anda:', 'NIM Binusian:', 'Fakultas:', 'Jurusan:',
                'Apakah Anda merupakan mahasiswa yang aktif dalam program Binus Mobility?',
                'Home Campus:', 'Seberapa sering Anda mengunjungi kantin Binus Anggrek dalam seminggu?',
                'saran_produk', 'saran_fasilitas', 'saran_kebersihan', 'saran_pelayanan',
                'Start time', 'Completion time']
likert_cols = [col for col in df_clean.columns if col not in exclude_cols]
```

```python
for col in likert_cols:
    df_clean[col] = df_clean[col].str.lower().map(likert_map)
```

```python
df_clean['Seberapa sering Anda mengunjungi kantin Binus Anggrek dalam seminggu?'] = \
    df_clean['Seberapa sering Anda mengunjungi kantin Binus Anggrek dalam seminggu?'].str.lower().map(freq_map)
```

```python
df_clean['Apakah Anda merupakan mahasiswa yang aktif dalam program Binus Mobility?'] = \
    df_clean['Apakah Anda merupakan mahasiswa yang aktif dalam program Binus Mobility?'].str.lower().map(bool_map)
```

# Data Visualization

## Distribusi Fakultas

```python
if 'Fakultas:' in df.columns:
    fakultas_count = df['Fakultas:'].value_counts()
    total = fakultas_count.sum()
    colors = sns.color_palette("Set1", n_colors=len(fakultas_count))
    explode = [0.05] * len(fakultas_count)

    plt.figure(figsize=(8, 8))
    wedges, texts = plt.pie(
        fakultas_count,
        startangle=90,
        colors=colors,
        explode=explode
    )

    # Buat label legend dengan jumlah + persentase
    labels = [
        f"{name} ({count} orang - {count/total:.1%})"
        for name, count in zip(fakultas_count.index, fakultas_count.values)
    ]

    plt.legend(wedges, labels, title="Fakultas", bbox_to_anchor=(1, 0.5), loc="center left", fontsize=9)
    plt.title("Distribusi Fakultas Responden")
    plt.tight_layout()
    plt.show()
```

## Distribusi Kepuasan Pelayanan di Kantin

```python
pelayanan_cols = [
    'Petugas kantin tanggap ketika menerima pesanan atau keluhan.',
    'Petugas kantin selalu ramah dan membantu ketika dibutuhkan.',
    'Pelayanan kantin yang cepat dan efisien saat pembayaran',
    'Waktu tunggu saat memesan makanan/minuman masih dapat ditoleransi.',
    'Metode pembayaran di kantin sudah memudahkan mahasiswa',
    'Proses pembayaran berjalan lancar dan tidak memakan waktu lama.',
    'Kantin menyediakan informasi terkait menu, dan harga dengan jelas.'
]
df_clean['rata_pelayanan'] = df_clean[pelayanan_cols].mean(axis=1)
```

```python
plt.figure(figsize=(4,4))
sns.boxplot(y=df_clean['rata_pelayanan'], color='#03A9F4')
plt.title("Distribusi Kepuasan Pelayanan")
plt.ylabel("Skor Rata-Rata")
plt.tight_layout()
plt.show()
```

Boxplot ini menggambarkan distribusi nilai rata-rata kepuasan pelayanan. Median berada di sekitar nilai 4 (setuju), dengan kuartil pertama sedikit di bawah 4 dan kuartil ketiga mendekati 4,5. Hal ini menunjukkan sebagian besar mahasiswa merasa puas (nilai tinggi) terhadap produk makanan kantin. Titik-titik (outlier) di sekitar 2 menunjukkan beberapa nilai rendah, tapi sebagian besar data berkumpul di atas rata-rata.

```python
print(f'Skewness: {skew(df_clean["rata_pelayanan"], bias = False)}')
print(f'Kurtosis: {kurtosis(df_clean["rata_pelayanan"], bias = False)}')
print(df_clean["rata_pelayanan"].describe())
```

## Distribusi Kepuasan Produk Kantin

```python
product_cols = [
    'Menu yang disediakan beragam dan sesuai dengan selera mahasiswa.',
    'Terdapat pilihan makanan/minuman sehat atau bergizi di kantin.',
    'Makanan/minuman di kantin disajikan dalam kondisi segar dan terjaga kualitasnya.',
    'Rasa makanan/minuman di kantin sesuai dengan ekspektasi.',
    'Harga makanan/minuman di kantin terjangkau bagi mahasiswa.',
    'Harga makanan/minuman sesuai dengan kualitas yang diberikan.'
]
df_clean['rata_produk'] = df_clean[product_cols].mean(axis=1)
```

```python
plt.figure(figsize=(4,4))
sns.boxplot(y=df_clean['rata_produk'], color='#03A9F4')
plt.title("Distribusi Kepuasan Produk")
plt.ylabel("Skor Rata-Rata")
plt.tight_layout()
plt.show()
```

Boxplot ini menggambarkan distribusi nilai rata-rata kepuasan produk. Median berada di sekitar nilai 4 (setuju), dengan kuartil pertama ~4 dan kuartil ketiga sedikit di atas 4. Hal ini menunjukkan sebagian besar mahasiswa merasa puas (nilai tinggi) terhadap produk makanan kantin. Titik-titik (outlier) di bawah 3 menunjukkan beberapa nilai rendah, tapi sebagian besar data berkumpul di atas rata-rata.

```python
print(f'Skewness: {skew(df_clean["rata_produk"], bias = False)}')
print(f'Kurtosis: {kurtosis(df_clean["rata_produk"], bias = False)}')
print(df_clean['rata_produk'].describe())
```

Nilai median (Q2) sebesar 4.00, dengan kuartil pertama (Q1) sebesar 3.67 dan
kuartil ketiga (Q3) sebesar 4.17, menunjukkan bahwa 50% dari responden memberikan skor kepuasan antara 3.67 hingga 4.17. Ini berarti persepsi mahasiswa terhadap produk kantin cenderung positif, dengan skor terkonsentrasi di area tinggi (di atas nilai 3.5).

Rentang interkuartil (IQR = Q3  Q1 = 0.5) tergolong sempit, yang menunjukkan bahwa variabilitas penilaian antar responden tidak terlalu besar dan kebanyakan mahasiswa memberikan penilaian yang konsisten di kisaran "setuju".

Nilai minimum sebesar 2.83 mengindikasikan adanya responden dengan tingkat kepuasan rendah, sementara nilai maksimum sebesar 5.00 menunjukkan bahwa ada juga responden yang memberikan skor kepuasan tertinggi.

Skewness = 0.04  Angka ini sangat mendekati nol, menunjukkan bahwa distribusi data hampir simetris, dengan sedikit kecenderungan miring ke kanan (positif). Artinya, lebih banyak responden memberi nilai tinggi, tetapi ada sebagian kecil yang memberi skor di bawah rata-rata.

Kurtosis = -0.05  Nilai ini < 3 menandakan bahwa distribusi tergolong platikurtik, yaitu lebih datar dibanding distribusi normal. Ini berarti data tidak memiliki banyak nilai ekstrem (outlier), dan mayoritas nilai terkonsentrasi di sekitar rata-rata.
